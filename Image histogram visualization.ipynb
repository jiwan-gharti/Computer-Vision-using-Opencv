```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/eiffel.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
hist = cv2.calcHist([image],[0],None,[256],[0,255])
plt.hist(image.ravel(),256,[0,255])
plt.show()
```

```python
colors = ('r','g','b')
for i, color in enumerate(colors):
    histogram2 = cv2.calcHist([image],[i],None,[256],[0,255])
    plt.plot(histogram2,color=color)
    plt.xlim([0,256])
plt.show()
```
